# La ecuacion de Schrödinger
Ecuacion postulada por Erwin Schrödinger 1925 la cual dicta el comportamiento de la onda probabilistica $\Psi$ que representa a una particula en Mecanica Cuantica.

La ecuacion fue quiza el resultado mas importante para el desarrollo de la Mecanica Cuantica, es el equivalnete a la segunda ley de Newton

$F = m a $ (fisica clasica)

$i \hbar \partial_t \Psi[x,t) = \frac{-\hbar^2}{2m} \partial^2_x \Psi[x,t) + V[x,t) \Psi[x,t) $

donde:
* $i$ unidad imaginaria
* $\hbar$ la constante de planck reducida, importante numero en este campo
* $\partial_t$ derivada parcial con respecto al tiempo
* $m$ la masa de la particula
* $\partial^2_x$ segunda derivada con respecto a la posicion $x$
* $V[x,t)$ funcion potencial bajo el cual se encuentra la particula
* $\Psi[x,t)$ funcion onda probabilistica, este representa a la particula

Podemos usar unidades naturales a los atomos, en lugar de kilogramos, o metros que nos inventamos, podemos usar unidades propias de la naturaleza. El tema de como representar las unidades y resultados sera tratado una vez tengamos dichos resultados.


# Solucion Independiente del Tiempo

$i\hbar \partial_t \Psi = \hat{H}\Psi$

se le llama solucion estacionaria, ya que el valor promedio o de expectacion
del Hamiltoniano es una constante atemporal

$ E \Psi = \hat{H}\Psi$

Resolviendo la primera ecuacion con E como un eigenvalue, osea que reemplaza al operador
por un escalar (real en este caso). Separaremos la dependencia temporal de la espacial.

$\Psi = \psi(x) \phi(t)$

$i\hbar \; d_t \phi = E \phi$

$  \frac{d \phi}{\phi} = \frac{E}{i\hbar} \;dt $

$ \ln{\phi} = \frac{-i E t}{\hbar} + A$

$ \phi = Ae^{-iEt/\hbar}$

```python
import numpy as np
import matplotlib.pyplot as plt
#hbar = 1

phi = lambda E,t: np.exp(-1j * E * t)
#phi[i](j) #energia i, con tiempo j
psi = lambda x: np.cos(x) 

t = np.linspace(0,10,100)
```

```python
for E in range(0,3):
    plt.plot(t,phi(E,t),label="$\phi({},t)$".format(E)) 
    
plt.legend()
plt.show()
```

Se llama estacionario por que la probabilidad (la cantidad real)

$\rho = \Psi^* \Psi = \psi^* \psi \; e^{(iE-iE)t/\hbar} = \psi^* \psi $
